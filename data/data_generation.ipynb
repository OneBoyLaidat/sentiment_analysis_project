```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Sample fintech complaint templates
templates = [
    "My transfer of 5000 to my mom failed but I was debited.",
    "The app keeps crashing when I try to pay my electricity bill.",
    "I haven't received the airtime I purchased since yesterday.",
    "Fast and reliable service, thank you Phlex!",
    "Why is my account restricted? I need access to my funds now!",
    "Great interface, but the OTP takes too long to arrive.",
    "Double debit on my Netflix subscription. Please refund.",
    "I love the new UI update, very smooth.",
    "Customer support hasn't responded to my ticket for 3 days.",
    "Transaction failed but my balance was deducted. This is frustrating."
]

categories = ["Transfer", "Bill Payment", "Account Issue", "General Feedback", "Airtime/Data"]

data = []
for i in range(1, 5001):
    date = datetime(2025, 8, 13) + timedelta(days=random.randint(0, 90))
    data.append({
        "ticket_id": f"PHX-{1000 + i}",
        "timestamp": date.strftime("%Y-%m-%d %H:%M:%S"),
        "category": random.choice(categories),
        "ticket_text": random.choice(templates)
    })

df = pd.DataFrame(data)
df.to_csv('complaints.csv', index=False)
print("Dataset 'complaints.csv' generated with 5,000 records.")
```
